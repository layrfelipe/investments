```python
import pandas as pd
```

```python
df = pd.read_excel('./data.xlsx')
df.head()
```

```python
small_caps = ['TEND3',
'AZUL4',
'DASA3',
'RECV3',
'UNIP6',
'AMBP3',
'FRAS3',
'SAPR11',
'RAPT4',
'TRIS3',
'MEGA3',
'STBP3',
'LEVE3',
'VVEO3',
'ORVR3',
'CVCB3',
'GGPS3',
'ENAT3',
'ALOS3',
'TGMA3',
'MYPK3',
'CSMG3',
'PTBL3',
'AALR3',
'BKBR3',
'TTEN3',
'BLAU3',
'MRVE3',
'LJQQ3',
'MBLY3',
'SIMH3']
```

```python
pontuaction_dict = {}
for ticker in small_caps:
    pontuaction_dict[ticker] = 0
```

```python
df = df[df['Papel'].isin(small_caps)]
```

```python
df_sorted_by_c5y = df.sort_values(by='Cresc. Rec.5a', ascending=False)
df_sorted_by_c5y.reset_index(inplace=True)
df_sorted_by_c5y.drop('index', axis=1, inplace=True)

for idx, row in df_sorted_by_c5y.iterrows():
    pontuaction_dict[row['Papel']] += (idx+1)
```

```python
df_sorted_by_roic = df.sort_values(by='ROIC', ascending=False)
df_sorted_by_roic.reset_index(inplace=True)
df_sorted_by_roic.drop('index', axis=1, inplace=True)

for idx, row in df_sorted_by_roic.iterrows():
    pontuaction_dict[row['Papel']] += (idx+1)
```

```python
df_sorted_by_debt = df.sort_values(by='Dív.Brut/ Patrim.', ascending=True)
df_sorted_by_debt.reset_index(inplace=True)
df_sorted_by_debt.drop('index', axis=1, inplace=True)

for idx, row in df_sorted_by_debt.iterrows():
    pontuaction_dict[row['Papel']] += (idx+1)
```

```python
df_sorted_by_ev_ebitda = df.sort_values(by='EV/EBITDA', ascending=True)
df_sorted_by_ev_ebitda.reset_index(inplace=True)
df_sorted_by_ev_ebitda.drop('index', axis=1, inplace=True)

for idx, row in df_sorted_by_ev_ebitda.iterrows():
    pontuaction_dict[row['Papel']] += (idx+1)
```

```python
final_df = pd.DataFrame([pontuaction_dict], index=['Pontos'])
final_df = final_df.T
```
